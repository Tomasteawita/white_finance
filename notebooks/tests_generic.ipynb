```python
import pandas as pd
```

```python
file_path = "../../withe-finance-ingest/Cuenta Corriente PESOS 29-07-25.xlsx"
# leo el excel y lo convierto en un dataframe
df = pd.read_excel(file_path, sheet_name=None)
# obtengo el nombre de la hoja
sheet_name = list(df.keys())[0]
# obtengo el dataframe de la hoja
df = df[sheet_name]
# impritmo el tipo de dato de cada columna
print("Tipos de datos de las columnas:")
print(df.dtypes)
# muestro las primeras filas del dataframe
df.head()
```

```python
# si liquida y opera son tipos de datos, de tipo fecha, continuo
if pd.api.types.is_datetime64_any_dtype(df['Liquida']) and pd.api.types.is_datetime64_any_dtype(df['Operado']):
    print("Las columnas 'Liquida' y 'Opera' son de tipo fecha.")
# Si las columnas anteriores tienen el formato YYYY-MM-DD, imprimo True, sino Imprimo False junto con el formato actual
if df['Liquida'].dt.strftime('%Y-%m-%d').equals(df['Liquida'].astype(str)) and df['Operado'].dt.strftime('%Y-%m-%d').equals(df['Operado'].astype(str)):
    print("Las columnas 'Liquida' y 'Opera' tienen el formato YYYY-MM-DD.")
else:
    print("Las columnas 'Liquida' y 'Opera' no tienen el formato YYYY-MM-DD.")
    print("Formato actual de 'Liquida':", df['Liquida'].dt.strftime('%Y-%m-%d').unique())
    print("Formato actual de 'Opera':", df['Operado'].dt.strftime('%Y-%m-%d').unique())

# Si los valores de columnas Comprobnante, Especie y Referencia no tienen espacio al principio o al final
if df['Comprobante'].str.strip().equals(df['Comprobante']) and df['Especie'].str.strip().equals(df['Especie']) and df['Referencia'].str.strip().equals(df['Referencia']):
    print("Las columnas 'Comprobante', 'Especie' y 'Referencia' no tienen espacios al principio o al final.")
else:
    print("Algunas columnas tienen espacios al principio o al final.")

# Si las columnas Numero es de topo entero, continuo
if pd.api.types.is_integer_dtype(df['Numero']):
    print("La columna 'Numero' es de tipo entero.")

# Si las columnas cantidad, Precio, Importe, y Saldo son de tipo float, continuo
if pd.api.types.is_float_dtype(df['Cantidad']) and pd.api.types.is_float_dtype(df['Precio']) and pd.api.types.is_float_dtype(df['Importe']) and pd.api.types.is_float_dtype(df['Saldo']):
    print("Las columnas 'Cantidad', 'Precio', 'Importe' y 'Saldo' son de tipo float.")

```
